#### Convert image to array

```python
from matplotlib import image
from matplotlib import pyplot
from numpy import asarray
```

```python
#Load the image using matplotlib
data = image.imread('opera_house.jpg')
```

```python
type(data)
```

```python
print(data.dtype)
print(data.shape)
```

```python
pyplot.imshow(data)
pyplot.show()
```
